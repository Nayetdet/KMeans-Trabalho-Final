```python
import numpy as np
import matplotlib.pyplot as plt

FILE = '149143370-149143378-001.txt'
```

```python
def display_histograms(initial_histogram_path, out_histogram_path):
    with open(initial_histogram_path) as file:
        initial_histogram = np.array([int(value) for value in file.readline().split()])

    with open(out_histogram_path) as file:
        out_histogram = np.array([int(value) for value in file.readline().split()])

    intersection = np.minimum(initial_histogram, out_histogram)
    fig, ax = plt.subplots(figsize=(8, 6))

    ax.bar(np.arange(len(initial_histogram)), initial_histogram, label='Imagem Inicial')
    ax.bar(np.arange(len(intersection)), intersection, color='red', label='Imagem Clusterizada', alpha=0.6)

    ax.set_title('Histograma da Imagem')
    ax.set_xlabel('Intensidade da Cor')
    ax.set_ylabel('FrequÃªncia')
    ax.legend()

    plt.tight_layout()
    plt.show()

display_histograms(f'histograms/initial/{FILE}', f'histograms/out/{FILE}')
```
