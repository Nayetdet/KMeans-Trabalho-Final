```python
import numpy as np
import matplotlib.pyplot as plt

FILE = '149143370-149143378-001.txt'
```

```python
def display_histograms(initial_histogram_path, out_histogram_path):
    initial_histogram = np.loadtxt(initial_histogram_path, dtype=int)
    out_histogram = np.loadtxt(out_histogram_path, dtype=int)

    intersection = np.minimum(initial_histogram, out_histogram)

    plt.figure(figsize=(8, 6))
    plt.bar(np.arange(len(initial_histogram)), initial_histogram, label='Imagem Inicial')
    plt.bar(np.arange(len(intersection)), intersection, color='red', label='Imagem Clusterizada')

    plt.title('Histograma da Imagem')
    plt.xlabel('Intensidade da Cor')
    plt.ylabel('FrequÃªncia')

    plt.legend()
    plt.tight_layout()
    plt.show()

display_histograms(f'histograms/initial/{FILE}', f'histograms/out/{FILE}')
```
